```python
# 원본 리스트 정의
a = [2, 3, 4, 5, 6]
print("a =", a)

# 새로운 빈 리스트 생성
rev_a = []

# for-in문과 pop() 메소드를 사용하여 리스트 뒤집기
for _ in range(len(a)):
    rev_a.append(a.pop())

print("rev_a =", rev_a)
```

```python
def find_most_frequent_largest(data_tuple):
    # 각 요소의 빈도수를 저장할 딕셔너리 생성
    frequency = {}

    # 튜플의 각 요소를 순회하며 빈도수 계산
    for item in data_tuple:
        if item in frequency:
            frequency[item] += 1
        else:
            frequency[item] = 1

    # 가장 높은 빈도수 찾기
    max_frequency = max(frequency.values())

    # 가장 높은 빈도수를 가진 요소들 찾기
    most_frequent_items = [item for item, freq in frequency.items() if freq == max_frequency]

    # 가장 높은 빈도수를 가진 요소들 중 가장 큰 값 찾기
    largest_most_frequent = max(most_frequent_items)

    return largest_most_frequent

# 첫 번째 예시
tuple1 = (1, 2, 5, 4, 3, 2, 1, 4, 7, 8, 9, 9, 3, 7, 3)
print("주어진 튜플은:", tuple1)
print("가장 많이 나타나는 원소는:", find_most_frequent_largest(tuple1))

# 두 번째 예시
tuple2 = (1, 2, 5, 4, 3, 2, 9, 4, 7, 8, 9, 9, 3, 7, 3)
print("\n주어진 튜플은:", tuple2)
print("가장 많이 나타나는 원소는:", find_most_frequent_largest(tuple2))
```

```python
# 메뉴와 가격을 딕셔너리로 정의
menu = {
    "Americano": 3000,
    "Ice Americano": 3500,
    "Cappuccino": 4000,
    "Cafe Latte": 4500,
    "Espresso": 3600
}

# 메뉴 출력
for key in menu:
    print("{:16s}  가격 : {:,}원".format(key, menu[key]))

# 사용자로부터 주문 받기
choice = input('위의 메뉴중 하나를 선택하세요: ')

# 주문한 메뉴가 메뉴판에 있는지 확인
if choice in menu.keys():
    print('{}는 {:,}원 입니다. 결제를 부탁합니다.'.format(choice, menu[choice]))
else:
    print('미안합니다. {}는 메뉴에 없습니다'.format(choice))
```
