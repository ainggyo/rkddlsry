```python

import pandas as pd
import matplotlib.pyplot as plt


raw_data = pd.read_csv('/content/survey_results_public.csv')


print(raw_data.info())


target_data = raw_data[['Age', 'Country', 'LanguageHaveWorkedWith', 'LearnCode']]


print(target_data['LearnCode'].head())


methods = target_data['LearnCode'].str.split(';')


exploded_methods = methods.explode()

size_by_methods = exploded_methods.value_counts()


size_by_methods.nlargest(10).plot.pie(figsize=(5,5),autopct='%1.0f%%')
plt.title('Top 10 Methods People Learn to Code')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
